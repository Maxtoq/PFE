```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow.keras.applications.xception import Xception, decode_predictions
```

```python
model = Xception(include_top=True, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)
model.summary()
```

```python
cat = cv2.imread("../../Cat03.png")
cat = np.expand_dims(cat, 0)
cat.shape
```

```python
y_hat = model.predict(cat)
```

```python
np.where(y_hat[0] > 0.5)
```

```python
y_hat.shape
```

```python
decode_predictions(y_hat, top=5)
```

```python
mix = cv2.imread("../../mix.jpg")
mix = np.expand_dims(mix, 0)
mix.shape
```

```python
y_hat = model.predict(mix)
decode_predictions(y_hat, top=5)
```

```python
rott = cv2.imread("../../rott.jpg")
rott = np.expand_dims(rott, 0)
rott.shape
```

```python
y_hat = model.predict(rott)
decode_predictions(y_hat, top=5)
```

```python
screw = cv2.imread("../../index.jfif")
screw = np.expand_dims(cv2.cvtColor(screw, cv2.COLOR_BGR2RGB), 0)
screw.shape
```

```python
y_hat = model.predict(cv2.cvtColor(screw[0], cv2.COLOR_BGR2RGB))
decode_predictions(y_hat, top=5)
```
