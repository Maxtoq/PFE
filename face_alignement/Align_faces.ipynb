```python
import sys
import cv2
import dlib
```

```python
detector = dlib.get_frontal_face_detector()

sp = dlib.shape_predictor("shape_predictor_5_face_landmarks.dat")

```

```python
img = dlib.load_rgb_image("pers.jpg")
window = dlib.image_window()

window.set_image(img)
dets = detector(img, 1)
num_faces = len(dets)
print(num_faces)
```

```python
vid = cv2.VideoCapture("033.mp4")
success,image = vid.read()
# print(success)
count =0 
while success:
    dets = detector(image, 0)
    faces = dlib.full_object_detections()
    for detection in dets:
        faces.append(sp(image, detection))
        print(detections)
    images = dlib.get_face_chips(image, faces, size=160, padding=0.25)
    #image = dlib.get_face_chip(image, faces[0])
    count2 =0
    for image in images:
        cv2.imwrite("Visage{},Frame{}.jpg" .format(count2 ,count), image)     # save frame as JPEG file 
        count2 += 1
#         print(success)
    success,image = vid.read()
    print('Sur la frame {} il y a {} visage'.format(count, len(dets)))
    count += 1
```
